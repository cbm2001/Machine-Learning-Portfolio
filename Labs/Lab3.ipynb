## PCA ##

```python
# creating a copy of the data
data_copy = pd.read_csv('vehicle.csv')

#data_copy['class'].value_counts()

data_copy['class'] = data_copy['class'].replace("car", 0)
data_copy['class'] = data_copy['class'].replace("van", 1)
data_copy['class'] = data_copy['class'].replace("bus", 2)
```

```python
import matplotlib.pyplot as plt
```

```python
# scaling the data using min-max scaler, except the target variable
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features = data_copy.drop('class', axis=1)
scaled_data = scaler.fit_transform(features)

# creating a dataframe of scaled features
scaled_data = pd.DataFrame(scaled_data, columns=features.columns)

# adding the target variable to the dataframe
scaled_data['class'] = data_copy['class']
```

```python
# plotting the class distribution
plt.figure(figsize=(10, 6))
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
scaled_data['class'].value_counts().plot(kind='bar')
plt.show()
```

```python
# checking for missing values
scaled_data.isnull().sum()

# removing the missing values
scaled_data = scaled_data.dropna()

scaled_data_attributes = scaled_data.drop('class', axis=1)
```

```python
# PCA for finding optimal number of components to be used
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=None)
pca.fit(scaled_data_attributes)
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

### - choosing elbow point as 7

```python
# PCA with 7 components
pca = PCA(n_components=7)
pca.fit(scaled_data_attributes)  # X is your data
scaled_data_attributes_pca = pca.transform(scaled_data_attributes)

# creating a dataframe of the PCA features
scaled_data_attributes_pca = pd.DataFrame(scaled_data_attributes_pca)
scaled_data_attributes_pca['class'] = scaled_data['class']

scaled_data_attributes_pca
```

### - shown above is the dataframe after PCA with 7 components. 7 components were chosen because the elbow point was at 7, i.e, the variance explained by number of components greater than 7 plateaued.

## NMF

```python
# trying NMF with 7 components
from sklearn.decomposition import NMF
nmf = NMF(n_components=7)
nmf.fit(scaled_data_attributes)
scaled_data_attributes_nmf = nmf.transform(scaled_data_attributes)

# creating a dataframe of the NMF features
scaled_data_attributes_nmf = pd.DataFrame(scaled_data_attributes_nmf)
scaled_data_attributes_nmf['class'] = scaled_data['class']

scaled_data_attributes_nmf
```

### - shown above is the dataframe after PCA with 7 components.

## Interpretability of the components

### - NMF

```python
# create a dataframe of the NMF components
nmf_components_df = pd.DataFrame(nmf.components_, columns=scaled_data_attributes.columns)

# print the top 5 features for each NMF component
for i in range(nmf.n_components_):
    print(f"Component {i+1}:")
    print(nmf_components_df.iloc[i].nlargest(5))
    print()
```

### - PCA

```python
# create a dataframe of the PCA components
pca_components_df = pd.DataFrame(pca.components_, columns=scaled_data_attributes.columns)

# print the top 5 features for each PCA component
for i in range(pca.n_components_):
    print(f"Component {i+1}:")
    print(pca_components_df.iloc[i].nlargest(5))
    print()
```

```python
# logistic regression on the original data
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split

labels = scaled_data['class']
features = scaled_data_attributes

# stratified test-train split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=42)

# logistic regression
logreg_1 = LogisticRegression()

# will use cross validation to evaluate the model, along with confusion matrix and classification report
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report

# cross validation
y_train_pred = cross_val_predict(logreg_1, X_train, y_train, cv=5)

# confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred))

# classification report
print("Classification Report:")
print(classification_report(y_train, y_train_pred))
```

```python
# logistic regression on the PCA data

labels = scaled_data_attributes_pca['class']
features = scaled_data_attributes_pca

# stratified test-train split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=42)

# logistic regression
logreg_2 = LogisticRegression()

# will use cross validation to evaluate the model, along with confusion matrix and classification report
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report

# cross validation
y_train_pred = cross_val_predict(logreg_2, X_train, y_train, cv=5)

# confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred))

# classification report
print("Classification Report:")
print(classification_report(y_train, y_train_pred))
```
